# Data Analysis

## I isolate/group the highest transactions of each cardholder by sorting the amount column in descending order.

```python
# initial imports
import pandas as pd
import numpy as np
import random
from sqlalchemy import create_engine


# create a connection to the database
engine = create_engine("postgresql://postgres:postgres@localhost:5432/fraud_detection")

# Write the SQL query
query = """
SELECT t.id, t.date, t.amount, t.card, t.id_merchant, m.id, m.name, m.id_merchant_category, c.name
FROM transaction t
LEFT JOIN merchants m
ON m.id=t.id_merchant
LEFT JOIN merchant_category c
ON m.id_merchant_category=c.id
GROUP by c.name, t.id, t.date, t.amount, t.card, t.id_merchant, m.id, m.name, m.id_merchant_category, c.name
ORDER BY t.amount;
"""

# Read the SQL query into a DataFrame
df = pd.read_sql(query, engine)
df.head()
```

# During the time period 7:00 a.m. to 9:00 a.m.

# These are the top 100 highest transactions during this time period



```python
# Write the SQL query
query = """
SELECT t.id, t.date, t.amount, t.card, t.id_merchant, m.id, m.name, m.id_merchant_category, c.name
FROM transaction t
LEFT JOIN merchants m
ON m.id=t.id_merchant
LEFT JOIN merchant_category c
ON m.id_merchant_category=c.id
where t.date::time between time '07:00:00' and '09:00:00'
GROUP by c.name, t.id, t.date, t.amount, t.card, t.id_merchant, m.id, m.name, m.id_merchant_category, c.name
ORDER BY t.amount DESC
LIMIT 100;
"""

# Read the SQL query into a DataFrame
seven_to_nine = pd.read_sql(query, engine)
seven_to_nine
```

## I have detected fraudulent or anomalous transactions. The average transaction amount from the top 100 highest transactions were 10 dollars to 20 dollars range. From index row (0 to 8), the average transactions amount were above 1000 dollars with 100 dollars as the lowest amount. Therefore, I concluded from index row 0 to 8 might be fraudulent transactions during 7 am to 9 am.

# Some fraudsters hack a credit card by making several small payments (generally less than 2.00),which are typicallt ignored by cardholders. By analyzing the amount of transactions less than 2.00 per cardholder, there's clear evidence that a credit card ahs been hacked. 


## According to the query below, from row 0 to row 10, show high evidence of a credit card being hack due to the duplicate transaction amount.


```python
# Write the SQL query
query = """
select t.card, t.amount, count(t.amount)
from transaction t
where t.amount < 2.00
GROUP BY t.card, t.amount
ORDER BY count(t.amount) desc;
"""

# Read the SQL query into a DataFrame
two = pd.read_sql(query, engine)
two.head(20)
```

# Below are the top 5 merchants prone to being hacked using small transactions

```python
# Write the SQL query
query = """
select t.id_merchant, count(t.id_merchant), t.amount, m.id, m.name
from transaction t
LEFT JOIN merchants m
ON m.id=t.id_merchant
GROUP BY t.id_merchant, t.amount, m.id, m.name
HAVING t.amount < 2
ORDER BY count desc;
"""

# Read the SQL query into a DataFrame
merchants = pd.read_sql(query, engine)
merchants.head()
```
